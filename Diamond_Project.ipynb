## **3) Data Cleaning:**

Data cleaning is the process of identifying and addressing the issues and errors in a dataset to ensure its accuracy, reliability and suitability for the analysis of the dataset. data cleaning includes handling the missing values, removing the duplicate values, correcting data types and normalising the data.
We can see the outliers in the data. We can remove those by finding the outliers and cleaning the NUL values.

```r
# Drop missing values of rows and columns

diamond_df <- diamond_data |>
              drop_na()

# Print dimension of clean dataset

dim(diamond_df)
```

## **4) Adaptation:**

Creating two additional columns based on the loaded dataset which are essential for further analysis of this data.

```r
# Creating vector for calculating volume of diamond

vec_1 <- c(diamond_df$length.mm..0..10.74..)
vec_2 <- c(diamond_df$width..mm..0..58.9..)
vec_3 <- c(diamond_df$depth)

volume_df <- c(vec_1*vec_2*vec_3)
```

```r
# Creating vector for calculating price per carat of diamond

vec_4 <- c(diamond_df$price)
vec_5 <- c(diamond_df$carat)

Price_per_Carat <- c(vec_4/vec_5)
```

```r
# Creating new data frame that includes volume and price per carat of diamond

Adapation_data <- cbind(diamond_df,volume_df, Price_per_Carat)

head(Adapation_data, 5)
```
